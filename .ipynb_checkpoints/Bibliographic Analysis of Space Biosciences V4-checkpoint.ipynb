```python
import pandas as pd
import numpy as np
import altair as alt
import matplotlib.pyplot as plt
# import seaborn as sns
import requests
import json
from pybliometrics import scopus
```

```python
import main
import Bibliography_Parse
import Abstract_Wordcloud
```

## Defined functions (only run, do not edit)

```python
# get all true positive data
BA_true_positives = pd.read_excel('Master Bibliography for AB_BA_SBE (1).xlsx', sheet_name="BA True Positives", skiprows=1)
BA_true_positives['PMID'] = BA_true_positives['PMID'].fillna(0)
BA_true_positives['ISBN'] = BA_true_positives['ISBN'].fillna(0)
BA_true_positives = BA_true_positives.astype({'PMID':'int', 'ISBN':'int'})


# add EID to BA True Positives Table
BA_true_positives_bib = Bibliography_Parse.clean_bibliography(BA_true_positives, Title=False)

# create instance to check query recall
BA_corpus = main.Corpus(BA_true_positives_bib)
```

## Only change the following cells

```python
# Query 1
query_1 = 'Title-Abs-Key("Bioastronautics" OR "space medicine" OR "medicine in space")'
BA_corpus.check_recall(query_1, view='STANDARD')
```

```python
# Query 2
query_2 = 'KEY((human OR biol*) AND spacefl*)'
BA_corpus.check_recall(query_2, view='STANDARD')
```

```python
# Query 3
query_3 = 'TITLE-ABS-KEY((microgravit* OR "artificial grav*" OR spaceflight) AND (human OR biomed* OR cardio* OR "human health*"))'
BA_corpus.check_recall(query_3, view='STANDARD')
```

```python
# Query 4
query_4 = 'TITLE-ABS-KEY((astronaut* OR (human AND spaceflight)) AND (health OR physiolog* OR psycholog* OR vestib* OR ocula*))'
BA_corpus.check_recall(query_4, view='STANDARD')
```

```python
#space biomedicine
query_5 = 'TITLE-ABS-KEY((medic* OR drugs OR *vestib*) AND (astronaut* OR (human AND spaceflight)))'
BA_corpus.check_recall(query_5, view='STANDARD')
```

```python
merged_query = 'TITLE-ABS-KEY(("bioastronautics" OR "space medicine" OR "medicine in space") OR ((*astronaut* OR human*) AND (spaceflight OR microgravit* OR "artificial grav" OR mars OR martian OR moon OR "*space station") AND ("life science*" OR health OR medic* OR physiolog* OR psycholog* OR vestib* OR ocula* OR cardio* OR cardiac OR nervous OR drug* OR biol* OR immun* OR lymph* OR respirat* OR digestiv* OR reproductive OR "functional resilience")))'
BA_corpus.check_recall(merged_query, view='STANDARD')
```
